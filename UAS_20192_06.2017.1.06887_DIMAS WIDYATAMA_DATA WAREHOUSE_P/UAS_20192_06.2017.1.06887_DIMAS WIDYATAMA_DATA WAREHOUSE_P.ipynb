```python
# import library untuk manipulasi dan analisis data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

# baca semua data di file csv dan jadikan dataframe kemudian tampilkan
dataset = pd.read_csv('mahasiswa.csv')
dataset
```

```python
# tampilkan semua nama kolom dan tipe datanya
dataset.dtypes
```

# Laporan 1

```python
# 1. Laporan banyaknya pendaftar/mahasiswa baru/lulusan mahasiswa seluruh perguruan tinggi
# temukan pendaftar/mahasiswa baru/lulusan mahasiswa seluruh perguruan tinggi dan hitung banyak kemunculannya
mhs = dataset['indikator'].value_counts().reset_index()

# ubah nama kolom kemudian tampilkan laporan
mhs.columns = ['indikator', 'count']
mhs
```

```python
# visualisasi laporan 1
# temukan pendaftar/mahasiswa baru/lulusan mahasiswa seluruh perguruan tinggi dan hitung banyak kemunculannya

# menentukan ukuran visualisasi (panjang=24, lebar=8)
plt.figure(figsize=(10,4) )
# visualisasi bar chart horizontal dengan sumbu x=cabang perpustakaan, y=banyak pendaftaran
plt.barh(mhs['indikator'], mhs['count'],edgecolor='black', linewidth=1)
# memberi judul visualisasi
plt.title("PENDAFTAR/MAHASISWA BARU/LULUSAN")
# memberi judul label sumbu x dan y
plt.ylabel('indikator')
plt.xlabel('Count')
# simpan visualisasi lalu tampilkan
plt.savefig('laporan_1.png')
plt.show()
```

# Laporan 2

```python
# 2. Laporan Rata â€“ Rata Banyaknya pendaftar/mahasiswa baru/lulusan mahasiswa laki-laki seluruh perguruan tinggi
# kelompokkan data berdasarkan tipe indikator
# kemudian cari rata - rata total pendaftar/mahasiswa baru/lulusan mahasiswa laki-laki

indikator = dataset.groupby(['indikator'])['jumlah_lelaki'].mean().reset_index()
indikator
```

```python
# visualisasi laporan 2
plt.figure(figsize=(10,4))
plt.barh(indikator['indikator'], indikator['jumlah_lelaki'],edgecolor='black', linewidth=1)
plt.title("PENDAFTAR/MAHASISWA BARU/LULUSAN MAHASISWA LAKI-LAKI")
plt.ylabel('INDIKATOR')
plt.xlabel('Jumlah pendaftar/mahasiswa baru/lulusan mahasiswa laki-laki')
plt.savefig('laporan_2.png')
plt.show()
```

# Laporan 3

```python
# 3. Laporan Rata â€“ Rata Banyaknya pendaftar/mahasiswa baru/lulusan mahasiswa perempuan tiap jenis univeritas
# kelompokkan data berdasarkan jenis universitas
# cari rata - rata total pendaftar/mahasiswa baru/lulusan mahasiswa perempuan tiap jenis universitas

avg_co = dataset.groupby(['jenis_universitas'])['jumlah_perempuan'].mean().reset_index()
avg_co
```

```python
# visualisasi laporan 3

plt.figure(figsize=(10,3))
plt.barh(avg_co['jenis_universitas'], avg_co['jumlah_perempuan'],edgecolor='black', linewidth=1)
plt.title("PENDAFTAR/MAHASISWA BARU/LULUSAN MAHASISWA ")
plt.ylabel('Jenis Universitas')
plt.xlabel('Jumlah pendaftar/mahasiswa baru/lulusan mahasiswa perempuan')
plt.savefig('Laporan_3.png')
plt.show()
```

# Laporan 4

```python
# 4. Laporan Persebaran Tingkatan Pendidikan

# temukan tingkatan pendidikan dan hitung banyak kemunculannya
tingkat = dataset['tingkat'].value_counts().reset_index()
# ubah nama kolom lalu tampilkan laporan
tingkat.columns = ['tingkat', 'count']
# urutkan rentang umur dari yang paling rendah
tingkat = tingkat.sort_values('tingkat').reset_index().drop('index', 1)
tingkat
```

```python
# visualisasi laporan 4 dengan pie chart

# deklarasi label untuk pie chart
pie_labels = tingkat['tingkat']
# deklarasi untuk nilai tiap pie
ar_count = tingkat['count']
# menentukan ukuran visualisasi (panjang=14, lebar=10)
fig, ax = plt.subplots(figsize=(11,5))
# buat pie chart dengan nilai persentase 2 angka dibelakang koma dan dimulai dari 90 derajat
ax.pie(ar_count, labels=pie_labels, autopct='%1.2f%%', startangle=90)
ax.set_title("Tingkat Pendidikan")
ax.axis('equal')
plt.savefig('laporan_4.png')
plt.show()
```

```python
# 5. Laporan rata-rata banyaknya perguruan tinggi tiap jenis_universitas
dataset.sort_values(by='jenis_universitas', ascending=True)
```

```python
# visualisasi laporan 5
# cari rata-rata banyaknya perguruan tinggi tiap jenis_universitas

labels = 'Perguruan Tingi Negeri ' , 'Perguruan Tinggi Swasta'
sizes = [43, 46]
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.savefig('laporan_5.png')
plt.show()
```
